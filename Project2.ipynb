```python
import pandas as pd
```

```python
df1=pd.read_csv('/content/drive/MyDrive/project/new-term-test-1-result.csv')
df1
```

```python
df2=pd.read_csv('/content/drive/MyDrive/project/new-term-test-2-result.csv')
df2
```

```python
merged_df = pd.merge(df1, df2, on="Registration Number")
merged_df = merged_df.drop(columns=['Name_y']).rename(columns={'Name_x': 'Name'})
merged_df
```

```python
merged_df['Best Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].max(axis=1)
merged_df['Average Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1)
merged_df
```

```python
merged_df = merged_df.drop(columns=['TT-1 Marks', 'TT-2 Marks'])
merged_df
```

```python
df3=pd.read_csv('/content/drive/MyDrive/project/attendance.csv')
df3
```

```python
final_df = pd.merge(merged_df, df3, on="Registration Number")
final_df['Final Marks'] = (final_df['Term_marks'] * 0.7) + final_df['Attendance_marks'] + final_df['Average Marks']

final_df.to_csv("final result.csv", index=True)
final_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = pd.read_csv('final result.csv')

final_marks = data['Final Marks'].values.reshape(-1, 1)

K_values = range(2, 6)
for K in K_values:
    kmeans = KMeans(n_clusters=K, random_state=42)
    data[f'Cluster_K{K}'] = kmeans.fit_predict(final_marks)


    plt.figure(figsize=(8, 6))
    for cluster in range(K):
        plt.scatter(data[data[f'Cluster_K{K}'] == cluster]['Final Marks'],
                    np.zeros_like(data[data[f'Cluster_K{K}'] == cluster]['Final Marks']),
                    label=f'Cluster {cluster + 1}')
    plt.title(f'K-Means Clustering with K={K}')
    plt.xlabel('Final Marks')
    plt.yticks([])
    plt.legend()
    plt.show()
```
